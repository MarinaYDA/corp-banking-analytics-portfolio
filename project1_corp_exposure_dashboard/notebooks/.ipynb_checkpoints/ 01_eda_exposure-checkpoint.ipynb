# Project 1 – Corporate Exposure & Concentration  
## 01 – Exploratory exposure analysis

Goals of this notebook:

- Load the `corp_exposure_snapshot.csv` dataset.
- Compute core exposure KPIs:
  - Total limits and outstanding exposure.
  - Exposure by group, sector, country, and rating_band.
- Identify top obligors / groups and overall concentration.
- Prepare a short summary to feed the dashboard and report.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", lambda x: f"{x:,.2f}")
%matplotlib inline

data_path = "../data/corp_exposure_snapshot.csv"
df = pd.read_csv(data_path, parse_dates=["origination_date", "maturity_date"])

df.shape, df.head()
```

```python
df.info()
df.describe(include="all").T
```

```python
# Core KPIs:
total_limit = df["limit_amount"].sum()
total_outstanding = df["outstanding_amount"].sum()
total_undrawn = df["undrawn_amount"].sum()

utilization = total_outstanding / total_limit

kpi = pd.DataFrame({
    "metric": ["Total limit", "Total outstanding", "Total undrawn", "Utilization"],
    "value": [total_limit, total_outstanding, total_undrawn, utilization]
})
kpi
```

```python
# Top groups:
group_exposure = (
    df.groupby("group_name")["outstanding_amount"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

group_exposure["share_pct"] = (group_exposure["outstanding_amount"] / total_outstanding * 100).round(2)
group_exposure.head(20)
```

```python
# Exposure by sector, country, rating_band:
sector_exposure = (
    df.groupby("sector")["outstanding_amount"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

country_exposure = (
    df.groupby("country")["outstanding_amount"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

rating_exposure = (
    df.groupby("rating_band")["outstanding_amount"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

sector_exposure, country_exposure, rating_exposure
```

```python
plt.figure(figsize=(8, 4))
plt.bar(sector_exposure["sector"], sector_exposure["outstanding_amount"] / 1e6)
plt.xticks(rotation=30, ha="right")
plt.ylabel("Outstanding exposure (USD millions)")
plt.title("Corporate exposure by sector")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(country_exposure["country"], country_exposure["outstanding_amount"] / 1e6)
plt.xticks(rotation=30, ha="right")
plt.ylabel("Outstanding exposure (USD millions)")
plt.title("Corporate exposure by country")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(rating_exposure["rating_band"], rating_exposure["outstanding_amount"]/1e6)
plt.ylabel("Outstanding exposure (USD millions)")
plt.title("Corporate exposure by rating band")
plt.tight_layout()
plt.show()
```

## 3. Portfolio overview – key findings

**Overall exposure**

- Total committed limit is about **$570.5M**, of which **$339.9M** is currently outstanding.
- That leaves roughly **$230.6M undrawn**, so the portfolio is **~60% utilized** – there is still meaningful headroom on existing facilities.

**Single-name concentration**

- The largest corporate group has around **$17.2M** outstanding (≈ **5.0%** of the total book).
- The top 20 groups each account for **~1.6–5.0%** of exposure, so concentration is visible but still reasonably diversified across large corporates rather than a single dominant name.

**Sector distribution**

- Exposure is led by **Manufacturing (~$69.7M)**, **TMT (~$66.4M)** and **Retail (~$64.1M)**.
- Mid-tier sectors include **Real Estate (~$41.4M)** and **Transport (~$36.9M)**, with smaller books in **Utilities (~$34.5M)** and **Energy (~$26.8M)**.
- Overall, the book is tilted toward real-economy corporates with a moderate share in cyclical sectors (Manufacturing / Retail / TMT).

**Country distribution**

- The largest country books are **JP (~$74.7M)** and **CN (~$50.3M)**, followed by **BR, IT, US, DE, MX, FR, UK**, and smaller exposures in **KZ** and **ES**.
- No single country exceeds **~22%** of total exposure, suggesting geographic diversification across Asia, Europe and the Americas.

**Credit quality (rating band)**

- The portfolio is dominated by **Investment Grade** corporates (≈ **$214.6M** outstanding).
- **Sub-IG** names account for ~**$88.5M**, and **Watchlist** exposure is limited to ~**$36.9M**.
- This mix is consistent with a **primarily investment-grade corporate book** with a controlled but meaningful share of higher-risk names for yield.

**Takeaway**

- Utilization, concentration, sector and rating splits are all consistent with a **large-corporate lending portfolio**: diversified by obligor and country, tilted to investment-grade names, and with manageable headroom on undrawn limits.
