# 01 – EDA & Rating Migration (Corporate Early-Warning Project)

In this notebook I explore the synthetic **corporate rating panel** created in
`00_generate_corp_rating_data.ipynb`:

- Understand rating distributions and portfolio composition.
- Analyse **rating migration** over time (upgrades, downgrades, stability).
- Compute simple **migration matrices** (rating_t → rating_t+1).
- Estimate **default rates by rating band**.
- Explore early-warning features (leverage, interest coverage, days past due)
  across ratings and default outcomes.

The output is an EDA baseline that sets up the early-warning model in
`02_early_warning_model.ipynb`.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)

data_dir = Path("../data")
panel_path = data_dir / "corp_rating_panel.csv"

panel_path
```

```python
df = pd.read_csv(panel_path, parse_dates=["as_of_date"])

df.head()
```

```python
df.shape, df.dtypes
```

# Basic sanity checks

```python
df["as_of_date"].min(), df["as_of_date"].max()
```

```python
df[["obligor_id", "obligor_name"]].nunique()
```

```python
df["rating_numeric"].describe()
```

```python
df["rating_band"].value_counts(normalize=True) * 100
```

```python
rating_counts = (
    df["rating_numeric"]
    .value_counts()
    .sort_index()
)

rating_counts
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
rating_counts.plot(kind="bar", ax=ax)
ax.set_title("Rating distribution (all periods)")
ax.set_xlabel("Internal rating (1=best, 10=worst)")
ax.set_ylabel("Count of obligor-quarters")
plt.tight_layout()
plt.show()
```

```python
band_counts = (
    df["rating_band"]
    .value_counts()
    .reindex(["Investment_Grade", "Sub_IG", "Watchlist", "Default"])
)

band_counts
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
band_counts.plot(kind="bar", ax=ax)
ax.set_title("Rating band distribution (all periods)")
ax.set_xlabel("Rating band")
ax.set_ylabel("Count of obligor-quarters")
plt.tight_layout()
plt.show()
```

```python
df = df.sort_values(["obligor_id", "as_of_date"])

# Current and next rating_numeric
df["next_rating"] = df.groupby("obligor_id")["rating_numeric"].shift(-1)
df["next_band"] = df.groupby("obligor_id")["rating_band"].shift(-1)

# Keep only rows where we have a next state
mig = df.dropna(subset=["next_rating"]).copy()
mig["next_rating"] = mig["next_rating"].astype(int)

mig.shape, mig.head()
```

```python
migration_matrix = pd.crosstab(
    mig["rating_numeric"],
    mig["next_rating"],
    normalize="index"
)

migration_matrix
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(migration_matrix, cmap="Blues")
plt.colorbar(label="Probability")
plt.xticks(
    ticks=range(len(migration_matrix.columns)), 
    labels=migration_matrix.columns
)
plt.yticks(
    ticks=range(len(migration_matrix.index)), 
    labels=migration_matrix.index
)
plt.xlabel("Next rating (t+1)")
plt.ylabel("Current rating (t)")
plt.title("Rating migration matrix (numeric ratings)")
plt.tight_layout()
plt.show()
```

```python
band_migration = pd.crosstab(
    mig["rating_band"],
    mig["next_band"],
    normalize="index"
).reindex(index=["Investment_Grade", "Sub_IG", "Watchlist", "Default"],
          columns=["Investment_Grade", "Sub_IG", "Watchlist", "Default"])

band_migration
```

```python
plt.figure(figsize=(6, 5))
plt.imshow(band_migration, cmap="Blues")
plt.colorbar(label="Probability")
plt.xticks(
    ticks=range(len(band_migration.columns)),
    labels=band_migration.columns,
    rotation=45
)
plt.yticks(
    ticks=range(len(band_migration.index)),
    labels=band_migration.index
)
plt.xlabel("Next band (t+1)")
plt.ylabel("Current band (t)")
plt.title("Rating band migration matrix")
plt.tight_layout()
plt.show()
```

```python
# Use final snapshot per obligor for a simple default rate view
last = (
    df.sort_values(["obligor_id", "as_of_date"])
      .groupby("obligor_id")
      .tail(1)
      .copy()
)

last["rating_band"].value_counts()
```

```python
default_rates = (
    last.groupby("rating_band")["is_default"]
        .mean()
        .reindex(["Investment_Grade", "Sub_IG", "Watchlist", "Default"])
)

default_rates
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
(default_rates * 100).plot(kind="bar", ax=ax)
ax.set_title("Default rate by rating band (final snapshot)")
ax.set_xlabel("Rating band")
ax.set_ylabel("Default rate (%)")
plt.tight_layout()
plt.show()
```

```python
leverage_stats = (
    df.groupby("rating_band")["leverage"]
      .describe()[["mean", "50%", "std"]]
      .rename(columns={"50%": "median"})
      .reindex(["Investment_Grade", "Sub_IG", "Watchlist", "Default"])
)

leverage_stats
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
leverage_stats["mean"].plot(kind="bar", ax=ax)
ax.set_title("Average leverage by rating band")
ax.set_xlabel("Rating band")
ax.set_ylabel("Debt / EBITDA")
plt.tight_layout()
plt.show()
```

```python
icr_stats = (
    df.groupby("rating_band")["interest_coverage"]
      .describe()[["mean", "50%", "std"]]
      .rename(columns={"50%": "median"})
      .reindex(["Investment_Grade", "Sub_IG", "Watchlist", "Default"])
)

icr_stats
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
icr_stats["mean"].plot(kind="bar", ax=ax)
ax.set_title("Average interest coverage by rating band")
ax.set_xlabel("Rating band")
ax.set_ylabel("Interest coverage (x)")
plt.tight_layout()
plt.show()
```

```python
# Using 'last' (final snapshot per obligor)
ew_stats = (
    last.groupby("is_default")[["leverage", "interest_coverage", "days_past_due"]]
        .mean()
        .rename(index={0: "Non-default", 1: "Default"})
)

ew_stats
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

ew_stats["leverage"].plot(kind="bar", ax=axes[0])
axes[0].set_title("Leverage")
axes[0].set_ylabel("Debt / EBITDA")

ew_stats["interest_coverage"].plot(kind="bar", ax=axes[1])
axes[1].set_title("Interest coverage")

ew_stats["days_past_due"].plot(kind="bar", ax=axes[2])
axes[2].set_title("Days past due")

for ax in axes:
    ax.set_xlabel("")
    ax.tick_params(axis="x", rotation=15)

plt.suptitle("Early-warning features by default status (final snapshot)")
plt.tight_layout()
plt.show()
```

## EDA & rating migration – key findings

- **Rating distribution**
  - The synthetic portfolio has most exposure in ratings **3–7**, with a tail of
    weaker names in **8–9** and a small fraction already in **Default (10)**.
  - By rating band, the book is primarily **Investment_Grade** with a meaningful
    **Sub_IG** and **Watchlist** tail, consistent with a wholesale corporate
    portfolio.

- **Rating migration**
  - The **numeric migration matrix** shows that most obligors are **stable**
    quarter-to-quarter (high probability on the diagonal).
  - Downgrades are slightly more likely than upgrades, especially from the
    weaker buckets (7–9), where we also see occasional jumps into **Default**.
  - At the **rating band** level, most Investment_Grade names stay IG, while
    Watchlist names have a visibly higher probability of moving into Default.

- **Default rates**
  - Using the final snapshot per obligor, default rates increase monotonically
    from **Investment_Grade → Sub_IG → Watchlist → Default**, as expected:
    rating bands are ordered by risk.
  - This validates that the synthetic data behaves like a realistic internal
    rating system.

- **Early-warning features**
  - **Leverage** increases and **interest coverage** decreases as we move from
    Investment_Grade to Watchlist and Default, matching intuition.
  - In the final snapshot, **defaulted obligors** show:
    - higher leverage,
    - lower interest coverage,
    - significantly higher days past due,
    compared with non-defaulted names.

These patterns are what we will exploit in the next notebook
(`02_early_warning_model.ipynb`) to build a baseline **early-warning model**
that flags high-risk obligors based on rating and early-warning features.
